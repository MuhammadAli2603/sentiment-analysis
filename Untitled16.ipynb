```python
# 1) Install / import everything
!pip install --quiet nltk

import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# 2) Download the NLTK resources we need
nltk.download('punkt')
nltk.download('stopwords')
# Download the 'punkt_tab' resource which is needed internally by word_tokenize
nltk.download('punkt_tab')

# %%
# 3) Load your DataFrame (assume you've already downloaded "IMDB Dataset.csv" into /content)
df = pd.read_csv('/content/IMDBDataset.csv')
df.head()

# %%
# 4) Define the preprocessing function
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # 1. Lowercase
    text = text.lower()
    # 2. Remove HTML tags and URLs
    text = re.sub(r'<.*?>', ' ', text)            # strips HTML tags
    text = re.sub(r'http\S+|www\S+', ' ', text)    # strips URLs
    # 3. Remove non-alphanumeric chars (keep spaces)
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    # 4. Tokenize (now that punkt is downloaded, this will not fail)
    tokens = word_tokenize(text)
    # 5. Remove stopwords and stem
    cleaned_tokens = [ps.stem(tok) for tok in tokens if tok not in stop_words]
    # 6. Rejoin into one string
    return ' '.join(cleaned_tokens)

# %%
# 5) Apply preprocessing (this can take a minute or two on 50k reviews)
df['cleaned_review'] = df['review'].apply(preprocess_text)

# Quick sanity‐check
df[['review', 'cleaned_review']].head(3)
```

```python
from sklearn.model_selection import train_test_split

# 1. Map "positive"/"negative" to 1/0
df['label'] = df['sentiment'].map({'positive': 1, 'negative': 0})

# 2. Define features (X) and target (y)
X = df['cleaned_review']  # the preprocessed text
y = df['label']           # 1 or 0

# 3. Train/Test split (85% train, 15% test), stratify to keep class balance
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.15,
    random_state=42,
    stratify=y
)

print(f"Train size: {X_train.shape[0]} samples")
print(f"Test size:  {X_test.shape[0]} samples")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    max_features=5000,     # keep the top 5k tokens/ngrams
    ngram_range=(1, 2),    # include unigrams and bigrams
    min_df=5               # ignore tokens that appear in fewer than 5 docs
)

# 1. Fit the vectorizer on the training set, transform X_train
X_train_tfidf = vectorizer.fit_transform(X_train)

# 2. Transform X_test (do NOT call fit again)
X_test_tfidf = vectorizer.transform(X_test)

print("TF–IDF shape on train:", X_train_tfidf.shape)  # e.g., (42500, 5000)
print("TF–IDF shape on test: ", X_test_tfidf.shape)   # e.g., (7500, 5000)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize with class_weight='balanced' to handle any small imbalance
lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)

# Train (fit) on training TF–IDF vectors
lr.fit(X_train_tfidf, y_train)

# Predict on the test set
y_pred = lr.predict(X_test_tfidf)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Classification report
print(classification_report(
    y_test, y_pred,
    target_names=['negative', 'positive']
))

# 2) Confusion matrix (heatmap)
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['pred neg', 'pred pos'],
    yticklabels=['actual neg', 'actual pos']
)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import joblib

# Save TF–IDF vectorizer
joblib.dump(vectorizer, 'tfidf_vectorizer.joblib')

# Save the trained Logistic Regression model
joblib.dump(lr, 'sentiment_lr_model.joblib')
```

```python
# If you restarted the Colab runtime and need to reload:
# vectorizer = joblib.load('tfidf_vectorizer.joblib')
# lr         = joblib.load('sentiment_lr_model.joblib')

def predict_sentiment(raw_text):
    # 1. Preprocess exactly as before
    cleaned = preprocess_text(raw_text)
    # 2. Vectorize with TF–IDF
    vec    = vectorizer.transform([cleaned])
    # 3. Get predicted label (0/1) and probability of positive class
    label  = lr.predict(vec)[0]
    prob   = lr.predict_proba(vec)[0][1]  # probability of class “1” (positive)
    sentiment = 'positive' if label == 1 else 'negative'
    return sentiment, prob

# Try it out:
samples = [
    "I absolutely loved this movie. It was fantastic!",
    "Worst movie ever—completely wasted my time."
]

for s in samples:
    sent, p = predict_sentiment(s)
    print(f"Text: {s}\n→ Sentiment: {sent} (prob {p:.2f})\n")
```
